```python
import numpy as np
import pandas as pd
from scipy.cluster import hierarchy
import matplotlib.pyplot as plt

eles = [[0.2, 0.1, 0.8],
        [0.7, 0.8, 0.8],
        [0.1, 0.1, 0.8],
        [0.8, 0.9, 0.8],
        [0.3, 0.9, 0.8],
        [0.2, 0.7, 0.8]]

frame = pd.DataFrame(np.array(eles), index=["A", "B", "C", 
"D", "E", "F"])

Z = hierarchy.linkage(frame, 'complete')
plt.figure()
plt.title("Hierarchical Clustering Dendrogram")
dn = hierarchy.dendrogram(Z, labels=frame.index)
plt.xlabel("data points with complete-linkage")
plt.ylabel("Euclidean Distance")

plt.show()
```
